# 1. Importing packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#2. Ozone 

```python
df_ozone = pd.read_csv('daily_ozone_2020.csv')
df_ozone.head(3)
```

```python
df_ozone.info()
```

```python
df_ozone.describe()
```

Dropping Unnecessary Columns


```python
df_ozone.drop('Parameter_Code',axis=1,inplace=True)
df_ozone.drop('POC',axis=1,inplace=True)
df_ozone.drop('Parameter_Name',axis=1,inplace=True)
df_ozone.drop('Sample_Duration',axis=1,inplace=True)
df_ozone.drop('Pollutant_Standard',axis=1,inplace=True)
df_ozone.drop('Method_Code',axis=1,inplace=True)
df_ozone.drop('Method_Name',axis=1,inplace=True)

df_ozone.head(3)
```


### Heatmap

```python
corrmat = df_ozone.corr()
fig = plt.figure(figsize = (20, 10))

sns.heatmap(corrmat, vmax = 1, square = True,annot=True,vmin=-1)
plt.show()
```

##Distribution Plot

AQI

```python
sns.distplot(df_ozone.AQI) 
plt.title("AQI distribution - Ozone")
```

Arithmetic_Mean

```python
sns.distplot(df_ozone.Arithmetic_Mean) 
plt.title("Arithmetic Mean Distribution - Ozone")
```

```python
sns.distplot(df_ozone.firstMaxValue) 
plt.title("First Max Value distribution - Ozone")
```

```python
sns.distplot(df_ozone.firstMaxHour) 
plt.title("First Max hour - Ozone")
```

```python
plt.figure(figsize=(50,8))
sns.boxplot(y = 'State_Name', x = 'AQI', data = df_ozone)
```

Arizona has high AQI with highest and the most outliers. This shows that the AQI in the state seems to be very inconsistent.  On the other hand, Alaska and Hawaii has a low AQI with no outlier. The AQI for ozone seems to be consistent.
